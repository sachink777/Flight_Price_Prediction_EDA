# Feature Engineering

```python
final_df['Day'] = final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month'] = final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year'] = final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
final_df.head()
```

```python
# final_df['Day']=final_df['Date_of_Journey'].apply(lambda x:x.split('/').[0])
# another logic
```

Applying feature engineering on Date of Journey column

```python
final_df['Day']=final_df['Day'].astype(int)
final_df['Month']=final_df['Month'].astype(int)
final_df['Year']=final_df['Year'].astype(int)
```

```python
final_df.info()
```

```python
final_df.drop(columns=['Date_of_Journey'],axis=1,inplace=True)
final_df.head()
```

Applying feature engineering on Arrival time column

```python
final_df['Arrival_Time'] = final_df['Arrival_Time'].str.split(' ').str[0]
```

```python
final_df.head()
```

```python
final_df['Arrival hour'] = final_df['Arrival_Time'].str.split(':').str[0]
final_df['Arrival_min'] = final_df['Arrival_Time'].apply(lambda x:x.split(':')[1])
```

```python
final_df.head(5)
```

```python
final_df.drop(columns=['Arrival_Time'],axis=1,inplace=True)
```

```python
final_df['Arrival hour']=final_df['Arrival hour'].astype(int)
final_df['Arrival_min']=final_df['Arrival_min'].astype(int)
```

```python
final_df.info()
```

```python
final_df['Dep_hr'] = final_df['Dep_Time'].str.split(':').str[0]
final_df['Dep_min'] = final_df['Dep_Time'].str.split(':').str[1]
```

```python
final_df.drop(columns=['Dep_Time'],axis=1,inplace=True)
```

```python
final_df['Dep_hr']=final_df['Dep_hr'].astype(int)
final_df['Dep_min']=final_df['Dep_min'].astype(int)
```

```python
final_df.head(2)
```

```python
final_df.info()
```

```python
final_df['Total_Stops'].unique()
```

```python
final_df[final_df['Total_Stops'].isnull()]
```

```python
final_df['Total_Stops']=final_df['Total_Stops'].map({'non-stop':0,'1 stop':1, '2 stops':2, '3 stops':3, '4 stops':4})
```

```python
final_df.drop(columns=['Route'],axis=1,inplace=True)
```

```python
final_df.head()
```

```python
final_df['Additional_Info'].unique()
```

```python
final_df['Duration_hr']=(final_df['Duration'].str.split('h').str[0])
final_df['Duration_min_teamp']=(final_df['Duration'].str.split('h').str[1])
final_df['Duration_min'] = (final_df['Duration_min_teamp']).str.split('m').str[0]
```

```python
final_df.head()
```

```python
final_df.drop(columns=['Duration_min_teamp'],axis=1,inplace=True)
final_df.drop(columns=['Duration'],axis=1,inplace=True)
```

```python
final_df.head()
```

```python
final_df['Duration_hr'] = final_df['Duration_hr'].astype(int)
final_df['Duration_hr'] = final_df['Duration_hr']*60
```

```python
final_df['Duration_min'] = final_df['Duration_min'].astype(int)
```

Dropped 2 records where duration was 5m (Outlier)

```python
final_df.drop(6474,axis=0,inplace=True)
final_df.drop(2660,axis=0,inplace=True)
```

```python
final_df['Duration_min']=final_df['Duration_min'].replace(to_replace='',value=0)
```

```python
final_df.head()
```

Finally we have converted all the categorical features into int

```python
final_df.info()
```

```python
final_df['Airline'].unique()
```

Label Encoding on the remaining Features

```python
from sklearn.preprocessing import LabelEncoder
laberencoder=LabelEncoder()
```

```python
final_df['Airline']=laberencoder.fit_transform(final_df['Airline'])
final_df['Source']=laberencoder.fit_transform(final_df['Source'])
final_df['Destination']=laberencoder.fit_transform(final_df['Destination'])
final_df['Additional_Info']=laberencoder.fit_transform(final_df['Additional_Info'])
```
